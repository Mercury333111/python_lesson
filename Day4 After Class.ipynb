#### 1.2 发送 GET 请求，获取响应

HTTP 协议 定义了两种方法。第一种是 GET 第二种是 POST

- GET 一般用于获取数据（目前）
- POST 一般用于提交数据 - 表单 登录框

__语法__ ：

```python
import requests
resp = requests.get(url, params=None, headers=None, cookies=None, auth=None, timeout=None)
resp = requests.get(url)
```

发送 GET 请求给 url。

- import requests 表示导入 requests 内部的所有功能，只有导入了这个包才能使用它提供的功能。
- resp = requests.get(url) 表示，从 requests 包里面找到 get 工具，并且让它去访问 url 网址，将获取到的结果储存在 resp 变量中

参数：

url – URL for the new Request object.  请求对象的URL地址。

params – (optional) Dictionary of GET Parameters to send with the Request.  (可选)使用字典作为GET请求的参数。

headers – (optional) Dictionary of HTTP Headers to send with the Request. （可选）使用字典作为GET请求的headers信息。

cookies – (optional) CookieJar object to send with the Request. （可选）使用CookieJar对象发送GET请求。

auth – (optional) AuthObject to enable Basic HTTP Auth.   （可选）AuthObject启用基本HTTP身份验证

timeout – (optional) Float describing the timeout of the request. （可选）请求超时的设置，用浮点数表示。


__示例__ ：

````python
import requests
resp = requests.get("http://weather.cma.cn/")
````

__查看响应内容__ ：

````python
print(resp.text)
````

可以看到出现了乱码。这是由于 requests 库在处理响应时，使用了错误的解码方式。
修复方法也很简单，我们将解码方式设置为 'utf-8' 即可。如果 'utf-8' 还是乱码，那就再试试 'gbk'、'gb2312'。

````python
resp.encoding = 'utf-8'
print(resp.text[0:3000])
````

#### 什么是 UTF-8、Unicode、GB2312、GBK

在讲这四个东西之前，大家需要先了解一下 ASCII：



ASCII（American Standard Code for Information Interchange，美国信息交换标准代码）是一种最早的字符编码标准，用于将英文字符和常用符号编码为7位二进制数字，共计128个字符，包括字母、数字、标点符号和控制字符等。它是计算机领域最常用的字符编码之一。

ASCII 既是字符集（charset），又是编码（encoding）。

- 字符集 charset：字符的集合  比如 ASCII 是 128 个字符的集合
- 编码 encoding：字符到二进制的双向映射  比如 ASCII 规定了英文字母和数字如何跟 8bit 二进制数互相转换

类比：

全班有 40 个学生：小明，小红，小光……这个叫 charset

全班有 40 个学生：

| 学号  | 姓名 |
|:----:|:----:|
| 1026 | 小明 |
| 1027 | 小红 |
| 1028 | 小光 |

这个叫 encoding

---

Unicode（统一码、万国码）是一个字符集，它包含了全球各种语言的字符和符号，共计137,994个字符，其中包括了拉丁字母、希腊字母、中文、日文、韩文等各种语言的字符。Unicode是当前主流的字符集（是 charset！），它可以用不同的编码方式来表示，其中最常用的是UTF-8编码。

UTF-8（8-bit Unicode Transformation Format）是Unicode字符集的一种编码（是 encoding！），它使用变长编码方式，可以用1-4个字节来表示不同的Unicode字符，具有兼容ASCII编码的特性，可以表示128个ASCII字符，同时也可以表示全球各种语言的字符和符号。UTF-8编码是当前互联网上最常用的字符编码方式。

GB2312（国家标准2312）是中国国家标准中的字符编码标准，它只包含了汉字和少量的符号字符，共计7,445个字符，其编码方式为双字节编码，其中包括了简体中文中的所有字符。GB2312编码逐渐被GBK编码所取代。

GBK（汉字内码扩展规范）是GB2312的扩展编码，它包含了更多的汉字字符和符号，共计21,692个字符，与GB2312编码兼容，同时也支持繁体中文和日文等语言的字符。GBK编码是目前在中国大陆地区广泛使用的字符编码方式。

- 编码过程：字符 --> 二进制
- 解码过程：二进制 --> 字符

我们如果用 GB2312 编码方式来解码（二进制 --> 字符）使用 UTF-8 编码的网页，或者反过来，那自然就会变成乱码了。

通过修改 resp.encoding = 'utf-8' 可以告诉 Python 要用什么编码方式来解码网页。

#### 1.3 处理异常情况

有时候，request.get 可能会失败。导致失败的原因有很多，例如：网络波动，对方服务器开小差，网址写错了等等。我们需要能够发现这样的错误。

语法

```python
try:
    resp = requests.get(url)
except Exception as e:
    print("出错了：", e)
```

这里的 try except 组合就构成了一种「执行一下试试看会不会出错，如果出错了，就抓住这个错误处理它一下」的模式。`as` 表示起名字。

但是，有的时候，虽然网址打错了，但只错了一点点。于是请求成功发出了，并且也接收到了响应。在这样的情况下，会得到 404 Not Found。
比起连请求都发不出去的级别，这只能算是一个小错误。requests 不会抱怨出现了这样的小错误，因此我们没办法在 except 分支当中抓到问题。

这时，我们就需要通过 if 条件语句来判断 resp.status_code 是不是为某些特殊值，从而分析请求有没有得到想要的回复。

* 没得到响应：报错，会自动进入 except 分支
* 得到响应：不报错
  * 得到的是有问题的响应，例如 404 Not Found。此时 resp.status_code 为 404
  * 得到的是正常的响应，200 OK。此时 resp.status_code 为 200

示例：

````python
try:
    resp = requests.get("http://a")
    if resp.status_code != 200:
        print("不正确的响应码：", resp.status_code, resp.reason)
except Exception as e:
    print("出错了：", e)
````

这里我们可以参考 HTTP Status Code 表。

2xx成功
这一类型的状态码，代表请求已成功被服务器接收、理解、并接受。

3xx重定向
这类状态码代表需要客户端采取进一步的操作才能完成请求。通常，这些状态码用来重定向，后续的请求地址（重定向目标）在本次响应的Location域中指明。客户端应当自动监测无限循环重定向（例如：A→B→C→……→A或A→A），因为这会导致服务器和客户端大量不必要的资源消耗。按照HTTP/1.0版规范的建议，浏览器不应自动访问超过5次的重定向。

4xx客户端错误
这类的状态码代表了客户端看起来可能发生了错误，妨碍了服务器的处理。除非响应的是一个HEAD请求，否则服务器就应该返回一个解释当前错误状况的实体，以及这是临时的还是永久性的状况。这些状态码适用于任何请求方法。浏览器应当向用户显示任何包含在此类错误响应中的实体内容。

5xx服务器错误
表示服务器无法完成明显有效的请求。这类状态码代表了服务器在处理请求的过程中有错误或者异常状态发生，也有可能是服务器意识到以当前的软硬件资源无法完成对请求的处理。除非这是一个HEAD请求，否则服务器应当包含一个解释当前错误状态以及这个状况是临时的还是永久的解释信息实体。浏览器应当向用户展示任何在当前响应中被包含的实体。

### 2. HTML 源码的简单处理 - 基于字符串操作

总体思路：

1. 确定要爬取的网页的结构

   你肯定是去爬取一堆网页。它们的结构存在共性也存在差异。需要找共性部分用于定位
   
2. 找到在目标信息所在位置的附近，出现次数少的（最好只出现一两次）、固定不变的内容（字符串的子串），用于定位切割
3. 重复第二步，从大切小，从小切到更小，直到只剩下目标信息附近的一小段字符串
4. 利用 slice 操作 `string[start:end:step]` 把我们需要的信息提取出来

abcdef   abc ab a cde ef bcdef bf(不是)

什么叫「子串」--> 一个字符串的一部分 substring

        定位子字符串

            不存在返回 -1

        从特定位置切开字符串

            切割 [start:end] 语法

            使用切割语法

            使用 find rfind 方法
        删除前后缀的空白字符
            lstrip - strip 剥

            strip

        组合「定位」和「切割」，提取有用的信息
            逐步缩小范围

````python
name = "比达拉崩巴斑得贝迪卜多比鲁翁比达拉崩巴斑得贝迪卜多比鲁翁比达拉崩巴斑得贝迪卜多比鲁翁"

last_found_index = 0
find_count = 0
while True:
    try:
        当前找到的下标 = name.index("比", last_found_index + 1)
        find_count = find_count + 1
        print("第", find_count, "次找到的位置：", 当前找到的下标)
        last_found_index = 当前找到的下标
    except:
        break

last_found_index = 0
find_count = 0
while True:
    current_found_index = name.find("比", last_found_index + 1)
    if current_found_index == -1:
        break
    find_count = find_count + 1
    print("第", find_count, "次找到的位置：", current_found_index)
    last_found_index = current_found_index
````
